** all about List **
List is an ordered, mutable, and indexed collection that can store multiple items of different data types in a single variable, allowing duplicates, and is defined using square brackets []

```python
l=["Tarun",1,2.5,6,7,8]
print(type(l))
print(l[0:3:2])
l[1:]="tarun"
print(l)

l.insert(2,"tomar")##insert tomar at index 2
print(l)

l.remove("t") ##first occurance of the element
print(l)

l.pop() ##last element will be remove
print(l)

index=l.index("tomar")
print(index)

print(l.count("a"))

l.sort()
print(l) ## for sorting

l.reverse()## for reversing
print(l)

l.clear() ## for removing all the items present in the list
print(l)
```

List Comprehension
basic syantax:- [expression for item in iterable]

```python
num=[]
for i in range(10):
  num.append(i**2)
print(num)


square=[i**2 for i in range(10)]
print(square)
```

with conditional logic:-[expression for item in iterable if condition]


```python
lst=[]
for i in range(10):
  if i%2==0:
    lst.append(i)
print(lst)

##this is list comprehension
even_num=[i for i in range(10) if i%2==0]
print(even_num)
```

NESTED LIST COMPREHENSION:-[expression for item1 in itereable for item2 in itereable2]

```python
lst1=[1,2,3,4]
lst2=['a','b','c','d']

pair=[[i,j] for i in lst1 for j in lst2]
print(pair)
```

list comprehension with function calls

```python
words=["hello","word","python","list"]
lengts=[len(word) for word in words]
print(lengts)
```
