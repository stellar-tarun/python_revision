Magic methods in python,also known as dunder method(double underscore methods),are speical methods that start and end with double underscores. These methods enable you to define the behaviour of objects for built-in operations,such as arithmatic operations,comparisons,and more.

magic methods are predefined methods in python that you can override to change the behaviour of your objects. Some common magic methods include:



```python
'''
->__init__: initializes a new instance of a class
->__str__: returns a string representation of an object
->__repr__ : returns an official stirng representation of an object
->__len__:returns the length of an object
->__getitem__ :gets an item from a container
->__setitem__: sets an item in a container
'''
```

```python
## basic methods 
class Person:
    def __init__(self,name,age):
        self.name=name
        self.age=age

    def __str__(self):
        return f"{self.name},{self.age} years old"
    
    def __repr__(self):
        return f"Person(name={self.name},age={self.age})"
person=Person("Tarun",21)
print(person)
print(repr(person))
```
