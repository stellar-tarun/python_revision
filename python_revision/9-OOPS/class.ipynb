### instance variable 

```python
class Dog:
    ## constructor
    def __init__(self,name,age):
        self.name=name ##instance variable
        self.age=age

##create a object

dog1=Dog("buddy",3)
print(dog1)
print(dog1.name)
print(dog1.age)
```

## instance methods


```python
class Dog:
    ## constructor
    def __init__(self,name,age):
        self.name=name ##instance variable
        self.age=age

    def bark(self): ##insatnce method (variety of categoray)
        print(f"{self.name} says woof")


dog1=Dog("buddy",3)
dog1.bark()
dog2=Dog("luccy",4)
dog2.bark()
```

### modeling a bank account

```python
##define a class for bank account

class BankAccount:
    def __init__(self,owner,balance=0):
        self.owner=owner
        self.balance=balance

    def deposit(self,amount):
        self.balance+=amount
        print(f"{amount} is deposited. New balance is {self.balance}")

    def withdraw(self,amount):
        if amount>self.balance:
            print("Insufficient funds!")
        else:
            self.balance-=amount
            print(f"{amount} is withdrawn. New balance is {self.balance}")

    def get_balance(self):
        return self.balance

## create a account

account=BankAccount("tarun",2000)
print(account.balance)
```

```python
account.deposit(100)
```

```python
account.withdraw(100)
```

```python
account.get_balance()
```
