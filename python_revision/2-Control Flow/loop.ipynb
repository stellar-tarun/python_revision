** nested loop**

```python
for i in range(5):
  for j in range(2):
    print(f"this is i {i} and this is j {j}")
```

```python
n=5
sum=0
for i in range(n+1):
  sum+=i
print(sum)
```

```python
n=5
sum=0
count=1
while(count<=n):
  sum+=count
  count+=1

print(sum)
```
