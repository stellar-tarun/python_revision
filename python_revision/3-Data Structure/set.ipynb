** all about set **
Set is an unordered, mutable, and unindexed collection that stores only unique items of different data types in a single variable, does not allow duplicates, and is defined using curly brackets {}.

```python
st={1,2,3,4,5,6}
print(st)
print(type(st))
```

```python
# st.remove(3)
# print(st)

st.discard(19) ## agar wo number nahi hoga tab bhi koi problem nahi hai error show nahi hoga unlike remvoe
print(st)

st.clear()## clear all the element
print(st)
```

set membership test

```python
my_set={1,2,3,4,5}
print(1 in my_set)
print(6 in my_set)
```

mathmatical operation

```python
set1={1,2,3,4,5}
set2={4,5,6,7,8}
print(set1.union(set2))
print(set1.intersection(set2))
print(set1.difference(set2))
print(set1.symmetric_difference(set2))
print(set1.isdisjoint(set2))
print(set1.issubset(set2))
print(set1.issuperset(set2))
```

counting unique words in text

```python
text="hey whassup writing this code at sunday midday "
words=text.split()
print(words)
unique_words=set(words)
print(unique_words)
print(len(unique_words))
```
