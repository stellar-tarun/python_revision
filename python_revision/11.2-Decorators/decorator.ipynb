## Decorators

Decorators are a powerful and flexible feature in python that allows you to modify the behaviour of a function or class method. They are commonly used to add functionality to funciton or methods without modifying thier actual code. This lesson covers the basics of decorators, including how to create and use them

## decorator

Decorator = a function that wraps another function to add extra behavior before and after it, using @ syntax. 

````python
@main_welcome ##course_introduction = main_welcome(course_introduction)
def course_introduction():
    print("This is an advanced python course")
````

decorators with arguments

````python
def repeat(n):
    def decorator(func):
        def wrapper(*args,**kwargs):
            for i in range(n):
                func(*args,**kwargs)
        return wrapper
    return decorator
````

````python
@repeat(3)
def say_hello():
    print("tarun")
````

````python
say_hello()
````

## flow of the above code

**Complete flow in one picture:**
```
repeat(3) 
→ n=3 stored
→ returns decorator

decorator(say_hello) 
→ func=say_hello stored
→ returns wrapper

say_hello = wrapper now

say_hello() called
→ wrapper() runs
→ loop 3 times
→ original say_hello() called 3 times
→ tarun tarun tarun
