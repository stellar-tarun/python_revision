### Inheritance in Python


inheritance is a fundamental concept in oops that allows a calss to inherit attributes and methods from another class. This lesson covers single inheritance and multiple inheritance demostrating how to create and use them in python

```python
class Car:
    def __init__(self,windows,doors,enginetype):
        self.windows=windows
        self.doors=doors
        self.enginetype=enginetype

    def drive(self):
        print("The person will drive the {self.enginetype} car")
        
```

```python
car1=Car(4,5,"petrol")

car1.drive()
```

Single Inheritance

```python
class Tesla(Car):
    def __init__(self,windows,doors,enginetype,is_self_driving):
        super().__init__(windows,doors,enginetype  ) ## calling parent class('''parent attribute''')
        self.is_self_driving=is_self_driving

    def self_driving(self):
        print(f"Tesla supports self driving : {self.is_self_driving} ")
```

```python
tesla1=Tesla(4,5,"electric",True)
tesla1.self_driving()
```

Multiple Inheritance

```python
##Base class 1 (parent class)
class Animal:
    def __init__(self,name):
        self.name=name
    def speak(self):
        print("subclass must implement this method")
##Base class 2 (parent class)
class Pet:
    def __init__(self,owner):
        self.owner=owner

##Derived class 
class Dog(Animal,Pet):
    def __init__(self,name,owner): 
        ##OR Animal.__init(name) is also another way by which we can access the parenet class attribute without using super bcoz just want name attribute form the animal function so for zero confusion we did in this way 
        Animal.__init__(self,name)
        Pet.__init__(self,owner)
    def speak(self):
        return f"{self.name} say woof"



## Crate an object

dog = Dog("Buddy","Tarun")
print(dog.speak())
        

```
