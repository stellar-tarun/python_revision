print("*", end="")
```
`end=""` means after printing `*` don't go to new line, stay on the same line. So all stars in the inner loop print on one line:
```
*****

print()
```
This prints nothing but moves to the **next line**. It runs after the inner loop finishes — so after one full row of stars is printed, it goes to next line.


**Visual of what happens row by row:**
```
j=0 → print("*", end="")  →  *

j=2 → print("*", end="")  →  ***
j=3 → print("*", end="")  →  ****
j=4 → print("*", end="")  →  *****

Write a program that calculates the factorial of a number input by the user using a while loop.

````python
a=int(input("Enter the number : "))
factorial=1
while(a>0):
    factorial=factorial*a
    a-=1

print(factorial)
````

solution -14
Write a program that checks if a number input by the user is a prime number using a for loop.

````python
number = int(input("Enter a number: "))
is_prime = True
if number <= 1:
    is_prime = False
else:
    for i in range(2, int(number ** 0.5) + 1):
        if number % i == 0:
            is_prime = False
            break
if is_prime:
    print(f"{number} is a prime number.")
else:
    print(f"{number} is not a prime number.")
````

solution-15

Write a program that prints the first n Fibonacci numbers, where n is input by the user.

````python
n = int(input("Enter the number of Fibonacci numbers to print: "))
a, b = 0, 1
count = 0
while count < n:
    print(a)
    a, b = b, a + b
    count += 1
````
