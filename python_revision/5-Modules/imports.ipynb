importing modules in python:Modules and Packages

```python
import math
print(math.sqrt(16))
print(math.pi)
```

```python
from math import sqrt,pi
print(sqrt(16))
print(pi)
```

```python
from math import *
print(sqrt(16))
print(pi)
```

importing custom package

```python
from package.maths import addition
addition(3,4)
```

```python
from package import maths
maths.addition(2,3)
maths.multiplication(3,4)
maths.divide(4,2)
```
