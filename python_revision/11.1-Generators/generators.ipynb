## Generators

Generators are a simpler way to create iterators.They use the yield keyword to produce a series of values lazily,which means they generate values on the fly and do not store them in memory.

```python
def square(n):
    for i in range(n):
        return i**2
```

```python
square(3) ## it will give 0 bcoz for the first iteration i=0 it will return so it will not move to i=1
```

for iterating to all the values one by one we have to use the yield keyword

```python
def square(n):
    for i in range(n):
        yield i**2 ## it will create a local variable and save the values over there and print them
```

```python
square(3)
```

```python
for i in range(3):
    print(i)
```

## EXample: Reading large files

generators are particularly useful for reaading large files becuase they allow you to process one line at a time without loading the entire file into memory.

```python
def read_large_file(file_path):
    with open(file_path,'r') as file:
        for line in file:
            yield line
```

```python
file_path='large_file.txt'

for line in read_large_file(file_path):
    print(line.strip())
```
