Operator Overloading

Operator overloading allows you to define teh behavior of operators(+,-,*,etc) for custom objects. You achieve this by overriding specific magic methods in your class

```python
'''
__add__(self,other):Adds two objects using the + operators
__sub__(self,other):subtracts two objects using - operators
__mul__(self,other):multiply two objects using * operators
__truediv__(self,other):divides two objects using / operators
__eq__(self,other):check if two objects are equal using == operators
__lt__(self,other):check if one objects is less then another using < operators
__gt__(self,other):check for greator
'''
```

```python
class Vector:
    def __init__(self,x,y):
        self.x=x
        self.y=y
    
    def __add__(self,other):
        return Vector(self.x+other.x,self.y+other.y)
    
    def __mul__(self,other):
        return Vector(self.x*other.x,self.y*other.y)
    
    def __sub__(self,other):
        return Vector(self.x-other.x,self.y-other.y)
    
    def __truediv__(self,other):
        return Vector(self.x/other.x,self.y/other.y)
    
    def __eq__(self,other):
        return Vector(self.x==other.x, self.y==other.y)
    
    def __str__(self): ## coz of this line Python will know how to display your Vector object as text.When you do print(v1+v2) it returns a Vector object, and Python just prints its memory address like: (<__main__.Vector object at 0x10f8a3d50>)
        return f"Vector({self.x}, {self.y})"
    
## create objects of the vector class
v1=Vector(2,3)
v2=Vector(15,19)

print(v1+v2) ## here just by using the operator symbol the function get call 
print(v1-v2)
print(v1*v2)
print(v1/v2)
print(v1==v2)
```
