** all about Dictionaries **
Dictionary is an ordered, mutable, and indexed collection that stores data in key-value pairs, does not allow duplicate keys, and is defined using curly brackets {}.

```python
dic={}
print(type(dic))
```

```python
dic=dict()
print(type(dic))
```

```python
dic={"name":"tarun","age":21,"height":6.2}
print(dic)
print(dic["name"])
print(dic.get("age")) ## accessing using get()
print(dic.get("last_name","not available"))## now here last_name key doesnt exist so the provide value get print
```

iterating over dic,keys,values, or items

```python
for keys in dic.keys(): ##keys
  print(keys)
```

```python
for values in dic.values(): ##values
  print(values)
```

```python
for keys,values in dic.items(): ##items
  print(f"{keys} = {values}")
```

Nested dic

```python
students={
    "student1":{
        "name":"tarun",
        "age":21
    },
    "student2":{
        "name":"tomar",
        "age":22
}
}
print(students)
```

Accessing nested dic

```python
print(students["student1"])
print(students["student1"]["name"])
print(students["student2"]["age"])
```

iterating over nested dic

```python
for student_id,student_info in students.items():
  print(f"{student_id}:{student_info}")
  for key,value in student_info.items():
    print(f"{key}:{value}")
```

Dictionary Comphrenhension

```python
d = {}
for i in range(6):
    d[i] = i**2  ## store in memory like this:- {0: 0} :- {0: 0,1: 1}
print(d)
```

```python
squares={x:x**2 for x in range(6)}
print(squares)
```

condition dicitionary comprehension

```python
evens={x:x**2 for x in range(10) if x%2==0}
print(evens)
```

Modifying the dic element

```python
print(dic)
dic["last_name"]="tomar" ##add a new key and value
print(dic)
dic["name"]="tarun tomar" ##replace
print(dic)

del dic["age"] ##delete a key and value
print(dic)
```

use a dic to count the frequency of elements in list

```python
num=[1,2,3,4,4,5,6,3,3,3]
freq={}

for i in num:
  if i in freq: ## this is how the counting actually working
    freq[i]+=1
  else:
    freq[i]=1
print(freq)
```

merge 2 dic into one

```python
dic1={"name":"tarun","age":21}
dic2={"height":6.2,"weight":80}
merge_dic={**dic1,**dic2}
print(merge_dic)
```
