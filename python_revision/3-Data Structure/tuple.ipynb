** all about tuples **
Tuple is an ordered, immutable, and indexed collection that can store multiple items of different data types in a single variable, allowing duplicates, and is defined using parentheses ().

```python
tuple=()
print(type(tuple))
```

tuples are immutable,but if you wanna change it,u just have to convert this first into list

```python
tup=('tarun',21,6.2)
convert=list(tup)
print(convert)
convert.append("tomar")
tup=tuple(convert)
print(convert)
```

```python
num=(0,1,2,3,4,5,1,1,1,6)
print(num)
print(num[0])

num[0:4]

num[::-1]

print(num.count(1))
print(num.index(1)) ##first occurance
```

packing and unpacking

```python
##packing
pack_tuple=1,"hello",3.14
print(pack_tuple)
```

```python
##unpacking
a,b,c=pack_tuple
print(a)
print(b)
print(c)
```

```python
##unpacking with *
num=(1,2,3,4,5,6)
first,*middle,last=num
print(first)
print(middle)
print(last)
```

tuple concatination

```python
num=(1,2,3,4,5,6)
mix_tuple=("Tarun",21,6.2)
concatination=num+mix_tuple
print(concatination)
```
