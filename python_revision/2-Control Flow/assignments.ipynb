print("*", end="")
```
`end=""` means after printing `*` don't go to new line, stay on the same line. So all stars in the inner loop print on one line:
```
*****

print()
```
This prints nothing but moves to the **next line**. It runs after the inner loop finishes — so after one full row of stars is printed, it goes to next line.


**Visual of what happens row by row:**
```
j=0 → print("*", end="")  →  *
j=1 → print("*", end="")  →  **
j=2 → print("*", end="")  →  ***
j=3 → print("*", end="")  →  ****
j=4 → print("*", end="")  →  *****
      print()              →  moves to next line
