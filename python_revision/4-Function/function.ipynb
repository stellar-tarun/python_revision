positional and keywords arguments

```python
def print_num(*args): ##positional argument (all the arguments that given over calling time)
 for num in args:
  print(num)

print_num(1,2,3,4,5,"tarun")
```

```python
def print_detail(**kwargs): ##keyword argument
  for i in kwargs.items():
    print(i)

print_detail(name="tarun",age=21,height=6.2)
```

```python
def print_detail(**kwargs): ##keyword argument
  for i,j in kwargs.items():
    print(i)

print_detail(name="tarun",age=21,height=6.2)
```

```python
def print_detail(**kwargs): ##keyword argument
  for i,j in kwargs.items():
    print(f"{i}:{j}")

print_detail(name="tarun",age=21,height=6.2)
```

combined positional and keyword arguments

```python
def print_details(*args,**kwargs):
  for value in args:
    print(f"positional argument : {value}")

  for key,value in kwargs.items():
    print(f"keyword arguments = {key}:{value}")

print_details(1,2,3,4,5,"tarun",last_name="tomar",age=21)
```

Example 1:temperature conversion

```python
def convert_temp(temp,unit):
  """this function convert temp between celsius and fahrenheit"""
  if unit=="c":
    return (temp*9/5)+32
  elif unit=="f":
    return (temp-32)*5/9
  else:
    return None
print(convert_temp(32,"c"))
```

Example 2: password strength checker

```python
def pass_strength(password):
  """this function check the strength of password"""
  if len(password)<8:
    return "weak"
  if not any(char.isdigit() for  char in password):
    return False
  if not any(char.isupper() for char in password):
    return False
  if not any(char.islower() for char in password):
    return False
  if not any(char in "!@#$%^&*()_+" for char in password):
    return False
  return True
pass_strength("Krisp@nto6113")
```

Example 3: Calculate the total cost of items in a shoping cart

```python
def calculate_total(cart):
  total_cost=0
  for item in cart:
    total_cost+=item['price'] * item['quantity']

  return total_cost
cart=[
    {'name':"apple",'price':0.5,'quantity':3},
    {'name':"banana",'price':0.25,'quantity':2},
    {'name':"orange",'price':0.75,'quantity':1}
]


calculate_total(cart)
```

Example 4:Check if a string is palindrome

```python
def is_palindrome(s):
  s=s.lower().replace(" ","")
  return s==s[::-1]
is_palindrome("racecar")
```

Lambda function
lambda function are small anonymous functions defined using lambda keyword.They can have any no of arguments but only one expression.They are commonly used for short operations or as arguments to higher-order functions.

```python
# lambda arguments:expression 
```

```python
addition =lambda a,b : a+b
print(type(addition))
print(addition(2,3))
```

```python
num=lambda a,b:a>b
print(num(2,3))
```

Map function
the map() function applies a given function to all items in an input list(or any other iterable) and returns a map object(an iterator).This is particularly useful for transforming data in a list comprehensively.

```python
def square(x):
  return x*x
square(10)
```

```python
numbers=[1,2,3,4,5,6]
list(map(square,numbers)) ##map(function_name,list_name)
```

```python
numbers=[1,2,3,4,5,6]
list(map(lambda x:x*x,numbers))
```

map multiple iterable

```python
num1=[1,2,3]
num2=[4,5,6]
add_num=list(map(lambda x,y:x+y,num1,num2))
print(add_num)
```

Filter function
the filter() function constructs an iterator from elements of an iterable for which a function returns true. it is used to filter out items from a list (or any other iterable) based on a condition

```python
def even(num):
  if num%2==0:
    return True
  else:
    return False
```

```python
lst=[1,2,3,4,5,6,7,8,9]
filter(even,lst) ##filter(function_name,list_name)
list(filter(even,lst))
```

```python
words=['APPLE','banana','orange']
upper_word = list(filter(lambda x: x != x.upper(), words))
print(upper_word)
```

filter with a lambda function and multiple conditions

```python
num=[1,2,3,4,5,6,7,8,9]
even_and_greater_than_5=list(filter(lambda x:x%2==0 and x>5,num))
print(even_and_greater_than_5)
```
