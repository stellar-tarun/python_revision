```python
### using the os module
import os
cwd=os.getcwd()
print(f"Current working directory is {cwd}")
```

```python
## create a new directory

new_directory="package"
os.mkdir(new_directory)
print(f"Directory '{new_directory}' created")
```

```python
##listing files and directories

items=os.listdir('.')
print(items)
```

```python
### joining paths

dir_name="folder"
file_name="file.txt"
full_path=os.path.join(os.getcwd(),dir_name,file_name,)
print(full_path) 



```

method:-1 checking if the path of the file is exists or not

```python
path='example.txt'

if os.path.exists(path):
    print(f"the path '{path}' exists")
else:
    print(f"the path '{path}' didnt exists")
    
```

method:-2 for checking if the path is exists or not


```python
import os

path='example.txt'
if os.path.isfile(path):
    print(f"the path '{path}' is a file")
elif os.path.isdir(path):
     print(f"the path '{path}' is a directory")
else:
      print(f"the path '{path}' is neither a file or a directory")
      
```

```python
## getting the absoulte path

relative_path='example.txt'
absolute_path=os.path.abspath(relative_path)
print(absolute_path)
```
