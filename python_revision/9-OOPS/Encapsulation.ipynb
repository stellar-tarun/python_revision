## Encapsulation


Encapsulation and abstraction are two fundamental principles of oops that help in designing robust,maintainable,and resuable code.Encapsulation involes bundling data and methods that operate on the data withing a single unit,while abstraction involves hiding comples implementation details and exposing only the necessary features.


Encapsulation is the concept of wrapping data (variables) and methods(functions) together as a single unit. it restricts direct access to some of the objects componets , which is a means of preventing accidental interface and misuse of the data.

```python
### Encapsulation with Getter and Setter Methods

## Public , protected , private and variables or access modifiers


class Person:
    def __init__(self,name,age):
        self.name=name ## public variables
        self.age=age ## ""
def  get_name(person):
    return person.name
person=Person("tarun",21)
print(f"this is {person.name} and my age is {person.age}")

get_name(person)
```

private variables

using __ underscore two times before the variable it make the variable private variable, such that now it cant be access outside the class  ex :- (self.name -> self.__name)

```python
class Person:
    def __init__(self,name,age):
        self.__name=name ## private variables
        self.__age=age ## ""
def  get_name(person):
    return person.__name

person=Person("tarun",21)
get_name(person)
```

Protected variable

using _ underscore one times before the variable it make the variable protected variable, such that now it cant be access outside the class but can be access by the derived class(child class )  ex :- (self.name -> self.__name)

```python
class Person:
    def __init__(self,name,age):
        self._name=name ## protected variables
        self._age=age ## ""

class Employee(Person):
    def __init__(self, name, age):
        super().__init__(name, age)

employee=Employee("Tarun",21)
print(employee._name)
```

Encapsulation with getter and setter

```python
class Person:
    def __init__(self,name,age):
        self.__name=name ## private variable
        self.__age=age ## ""ArithmeticError

    ## getter method for name
    def get_name(self):
        return self.__name
    
    def get_age(self):
        return self.__age
    
    ## setter for name
    def set_name(self,name):
        self.__name=name

    ## setter method for age(for changing )
    def set_age(self,age):
        if age>0:
            self.__age=age
        else:
            print("Age cannot be negative")

person=Person("Tarun",21)

##access using getter method
print(person.get_age())
print(person.get_name())

## setting the value using setter method
person.set_age(20)
print(person.get_age())

person.set_age(-5)
```
