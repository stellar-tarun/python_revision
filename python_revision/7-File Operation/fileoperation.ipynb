# File Operation- Read and Write Files

File handling is a crucial part of any programming language.Python provides built-in functions and methods to read and write to files, both text and binary.This lesson will cover the basics of file handling, including reading and writing text files and binary files. (r,w,a)

```python
##Read a while File

with open ('example.txt','r') as file:
    content=file.read() ## now it will read all the content from the example txt file
    print(content)
```

using for loop

```python
with open ('example.txt','r') as file:
    for line in file:
        print(line)
```

using strip function (upar wale me new line ka space aa ja raha hai leking strip ka use karke wo nahi aayega)

```python
with open ('example.txt','r') as file:
    for line in file:
        print(line.strip()) ## strip() removes the newline character
```

```python
##Writing a file(overwriting)

with open ('example.txt','w') as file:
    file.write('Hello world\n')
    file.write('this is new line\n')
```

writing a file without overwriting

```python
with open ('example.txt','a') as file:
    file.write("append operation taking place\n")
```

Or using function

```python
def count_text_file(file_path):
    with open( file_path,'r') as file:
        lines=file.readlines()
        line_count=len(lines)
        word_count=sum(len(line.split()) for line in lines) 
        char_count=sum(len(line) for line in lines)
    return line_count,word_count,char_count

file_path='example.txt'
lines,words,characters=count_text_file(file_path)
print(f'Lines: {lines}, Words: {words} , char: {characters}')
```

with seek()

```python
with open('example.txt','w+')as file:
     file.write("hello world\n")
     file.write("this is a new line\n")


## move the file cursor to the begining
     file.seek(0) ## 0 index

     ## read the content of the file
     content=file.read()
     print(content)
```
